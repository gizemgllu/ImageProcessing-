```python
import numpy as np
import matplotlib.pyplot as plt

img = np.ones((1000,1000,3))

point1 = (300,100)
point2 = (100,100)
point3 = (100,300)
point4 = (300,300)

def MoveObject(move_x, move_y):

    x1 = point1[0] + move_x
    y1 = point1[1] + move_y
    x2 = point2[0] + move_x
    y2 = point2[1] + move_y
    x3 = point3[0] + move_x
    y3 = point3[1] + move_y
    x4 = point4[0] + move_x
    y4 = point4[1] + move_y
    
    # connected points
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
def ScaleObject(scale_x, scale_y):
    # Scaling
    x1 = point1[0] * scale_x
    y1 = point1[1] * scale_y
    x2 = point2[0] * scale_x
    y2 = point2[1] * scale_y
    x3 = point3[0] * scale_x
    y3 = point3[1] * scale_y
    x4 = point4[0] * scale_x
    y4 = point4[1] * scale_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)

def RotateObject(angle):
  
    theta = np.radians(angle)
    
    rotation_matrix = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    
    p1_rotated = np.dot(rotation_matrix, [point1[0], point1[1]])
    p2_rotated = np.dot(rotation_matrix, [point2[0], point2[1]])
    p3_rotated = np.dot(rotation_matrix, [point3[0], point3[1]])
    p4_rotated = np.dot(rotation_matrix, [point4[0], point4[1]])
    
    x5 = [p1_rotated[0], p2_rotated[0], p3_rotated[0], p4_rotated[0], p1_rotated[0]]
    y5 = [p1_rotated[1], p2_rotated[1], p3_rotated[1], p4_rotated[1], p1_rotated[1]]
    plt.plot(x5, y5)

plt.figure(figsize=(10,8), dpi=100, facecolor='w')
plt.imshow(img)

MoveObject(300, 0)     
ScaleObject(2, 2)      
RotateObject(15)      

plt.show()
```
