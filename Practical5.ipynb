Create 2 different blending mode functions (for example, "difference" or "hard_light") based on the last 3 digits of your student ID number. The formulas from the lecture must be used to complete the task. The list of variants is available in ORTUS in the 7th practical lecture task description.

My variant is 075:
Mode 1: Dodge / color dodge 
Mode2: Overlay 


```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def color_dodge(A, B):
  
    return np.clip(A / (1 - B + 1e-5), 0, 1)

def overlay(A, B):
    return np.where(B <= 0.5, 2 * A * B, 1 - 2 * (1 - A) * (1 - B))

def load_image(image_path, size=(500, 500)):
    image = Image.open(image_path).resize(size)
    return np.asarray(image) / 255.0  

image_A_path = 'bduck.jpg.webp'  
image_B_path = 'kugu.jpg'  
image_A = load_image(image_A_path)
image_B = load_image(image_B_path)

assert image_A.shape == image_B.shape, "Images have to be same size!"

result_color_dodge = color_dodge(image_A, image_B)
result_overlay = overlay(image_A, image_B)


def display(input1, input2, result, mode_name):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    axs[0].imshow(input1)
    axs[0].set_title('Input Image 1')
    axs[0].axis('off')
    axs[1].imshow(input2)
    axs[1].set_title('Input Image 2')
    axs[1].axis('off')
    axs[2].imshow(result)
    axs[2].set_title(f'Result ({mode_name})')
    axs[2].axis('off')
    plt.tight_layout()
    plt.show()

display(image_A, image_B, result_color_dodge, 'Color Dodge')
display(image_A, image_B, result_overlay, 'Overlay')
```
