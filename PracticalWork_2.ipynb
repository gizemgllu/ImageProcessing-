```python
import numpy as np               
import matplotlib.pyplot as plt  

img = np.ones((600,600,3))

# Five control points 
def DrawBezier(x0, y0, x1, y1, x2, y2, x3, y3, x4, y4):
   
    ts = 0.001
    
    t = 0
    while t <= 1.0:
        
       
        B0 = (1-t)**4
        B1 = 4*(1-t)**3 * t
        B2 = 6*(1-t)**2 * t**2
        B3 = 4*(1-t) * t**3
        B4 = t**4
        
        x = int(x0*B0 + x1*B1 + x2*B2 + x3*B3 + x4*B4)
        y = int(y0*B0 + y1*B1 + y2*B2 + y3*B3 + y4*B4)
        
       
        if 0 <= x < 600 and 0 <= y < 600:
            img[x,y] = (1,0,0)
        
  
        t = t + ts
    
    return


DrawBezier(300, 380, 300, 50, 150, 350, 300, 500, 400, 300)

plt.figure(figsize=(10,8), dpi=100, facecolor='w')
plt.imshow(img)
plt.show()
```
