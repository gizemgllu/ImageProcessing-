During class you will be given the first half of the Bresenham algorithm that can draw straight horizontal lines (where dx>dy), complete this algorithm so that it could draw straight vertical lines (where dy>dx). 
Use the Bresenham algorithm to draw the last 3 digits of your student ID number using multiple straight lines (for example, if your student ID number is 230ADB075, then you have to draw the digits 075).

```python
#Importing all libraries

import matplotlib.pyplot as plt

import numpy as np

import PIL

import cv2

```

```python
import numpy as np                 
import matplotlib.pyplot as plt    


img = np.ones((600, 600, 3))  # White background


def DrawLine(x1, y1, x2, y2):

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

# Checking direction neg or postv.

    if x1 < x2:
        xs = 1
    else:
        xs = -1
        
    if y1 < y2:
        ys = 1
    else:
        ys = -1

 # starting points
    x = x1
    y = y1

 # horizontal or more vertical
    if dx > dy:
        p = 2 * dy - dx
        while x != x2:
            x = x + xs
            if p > 0:
                y = y + ys
                p = p + 2 * dy - 2 * dx
            else:
                p = p + 2 * dy
            img[y, x] = 0  


    else:
        p = 2 * dx - dy
        while y != y2:
            y = y + ys
            if p > 0:
                x = x + xs
                p = p + 2 * dx - 2 * dy
            else:
                p = p + 2 * dx
            img[y, x] = 0  

    return


## For my variant: 075

DrawLine(50, 100, 50, 200) 
DrawLine(50, 200, 100, 200)  
DrawLine(100, 200, 100, 100)  
DrawLine(100, 100, 50, 100)  


DrawLine(150, 100, 200, 100)  
DrawLine(200, 100, 170, 200)  


DrawLine(250, 100, 300, 100)  
DrawLine(250, 100, 250, 150) 
DrawLine(250, 150, 300, 150)  
DrawLine(300, 150, 300, 200) 
DrawLine(250, 200, 300, 200) 


## To show output
plt.imshow(img)
plt.show()
```
